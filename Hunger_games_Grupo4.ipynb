- Busqueda de outliers --> regla intercuartil


```
# Calculamos iqr para todas las variables

dataset = dataframe.iloc[:, :-1]
describe_df = dataset.describe()
iqr = describe_df.loc["75%"] - describe_df.loc["25%"]
iqr
```

```
# Calculamos los limites
upper_bound = describe_df.loc["75%"]+1.5*iqr #superior
lower_bound = describe_df.loc["25%"]-1.5*iqr #inferior
```

```
# Filtramos con pandas para ver los outliers
# Encontramos los outliers
bool_index = dataset > upper_bound
outliers = bool_index.sum(axis=1)
bool_index2 = outliers > 0
dataset.loc[bool_index2]
```

```
# Encontramos que variable produce el outlier
bool_index.loc[bool_index2]
```

```
# Encontramos los outliers
bool_index = dataset < lower_bound
outliers = bool_index.sum(axis=1)
bool_index2 = outliers > 0
dataset.loc[bool_index2]
```

```
bool_index.loc[bool_index2] # Variable produce el outlier
```

**3. ARBOL DE DECISION**

```
np.random.seed(123) 
```

```
X = dataframe.iloc[:,:-1]
Y = dataframe.iloc[:,-1:]
X_train, X_test, Y_train, Y_test = train_test_split(X, Y,stratify=Y)
```

```
Y.value_counts()/len(Y)
```

```
Y_test.value_counts()/len(Y_test)
```

Notamos que las proporciones de muestras que tengo de cada clase en el dataset son muy diferentes. Esto va a suponer que previamente podamos intuir que nuestros modelos no van a funcionar demasiado bien.

```
model = DecisionTreeClassifier(criterion="entropy")
model.fit(X_train, Y_train)
```

```
# Vemos gráficamente el arbol de decisión que se ha calculado:
figura, ax = plt.subplots(figsize=(70, 60))
plot_tree(model, feature_names= X.columns, fontsize=30)
plt.show()
```

- EVALUAMOS EL MODELO

```
# Vemos que el entrenamiento consigue un 100%
model.score(X_train, Y_train)*100
```

```
# Vemos que para el test:
model.score(X_test, Y_test)*100
```

```
# Evaluciacion 
f1_score(Y_test, model.predict(X_test), average='micro')
```

- Analisis de resultados

El algoritmo ha hecho que encaje perfectamente (100%) para el dataset de entrenamiento, dado que la división es ideal. 

Sin embargo, cuando lo hacemos en el siguiente conjunto de datos, vemos que siendo del 70% no es una división perfecta.
Obtenemos asi, overfiting.

El algoritmo clasifica muy bien el conjunto de entrenamiento pero para un conjunto nuevo como es el de test no funciona excesivamente bien. 

Ahora probaremos diferentes hiperparámetros del modelo para ver si podemos conseguir un mejor resultado.

**4. ARBOL DE DECISION CON HIPERPARAMETROS**

```
scores = []
modelos = []
for depth in [5,10,None]:
  for sample_split in [2,5,10]:
    for leaf in [1,5,10]:
      model = DecisionTreeClassifier(criterion="entropy", max_depth = depth, min_samples_split=sample_split,min_samples_leaf=leaf)
      model.fit(X_train, Y_train) 
      modelos.append(model)
      scores.append(f1_score(Y_test, model.predict(X_test), average='micro'))

    #Evaluamos el modelo
    #print("--------------------------------------------------------------------------")
    #print(f"Decision Tree con los hiperparámetros: \nMax_Depth = {depth} \nMin_Samples_Split = {sample_split}\n")
    #print(f"Score Train: {model.score(X_train, Y_train)}")
    #print(f"Score Test: {model.score(X_test, Y_test)}")
    #print(f"F1 Score: {f1_score(Y_test, model.predict(X_test), average='micro')}")

best = np.array(scores).argmax()
print("El mejor árbol de decisión es el conseguido por el modelo: \n")
print(modelos[best])
print("El score obtenido es de: \n")
print(scores[best])
```

- Analisis de resultados

 Observamos que realizando el arbol de decisión variando los hiperparámetros comparado con el inicial, conseguimos aumentar un poco el score del test. Y vemos que disminuye el entrenamiento, generalizando más dado que en el proceso anterior nos daba un 100%, de manera que al cambiar los cortes realizados en el conjunto de entrenamiento hemos mejorado los resultados en el test.

**5. BAGGING**

```
np.random.seed(123) 
```

```
base_estimator = LogisticRegression(max_iter=300)
model = BaggingClassifier(base_estimator, n_estimators=2000, max_samples=100, max_features=5)
```

```
model.fit(X_train, Y_train)
```

```
model.score(X_test, Y_test)
```

```
base_estimator2 = DecisionTreeClassifier(criterion="entropy", max_depth=10, min_samples_split=10) #Parámetros que mejor nos han funcionado antes
model2 = BaggingClassifier(base_estimator2, n_estimators=2000, max_features=4)
```

```
model2.fit(X_train, Y_train)
```

```
model2.score(X_test, Y_test)
```

Hemos mejorado un poco los resultados que teníamos, a pesar que al estar la base de datos mal distribuida respecto a cada clase no se puede llegar a conseguir scores muy altos. A pesar de esto, lo que si vemos es una gran diferencia entre el score obtenido por tanto por el decision tree como con el random forest (bagging de decision tree) con respecto al obtenido con el bagging en la regresión logística, que es bastante peor generalmente.
