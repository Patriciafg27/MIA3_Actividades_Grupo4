```python
import random
def segundos_a_segundos_minutos_y_horas(segundos):
    horas = int(segundos / 60 / 60)
    segundos -= horas*60*60
    minutos = int(segundos/60)
    segundos -= minutos*60
    return f"{horas:02d}:{minutos:02d}:{segundos:02d}"
```

```python
class Joker:
    
    def __init__(self, KO, arma, puntos):
        self.ko = KO
        self.arma = arma
        self.puntos = puntos
    
    def puntuacion(self,daño):
          self.puntos -= daño
          self.ko = False
          if self.puntos >0:
            print("Nivel de vida del Joker:", self.puntos,"\n\n")
          else:
            print("El Joker...ha...muerto.\n")

    def estado_armas(self):
        if(self.arma == True):
            print("Joker ha sido dañado por un PEM y esta en shok. Necesitará 10 segundos para recuperarse")
        else:
            print("Sigue intentandolo! <Yo no soy un monstruo, sólo voy un paso por delante>, dice Joker!")
```

```python
class Joker_Villain:
    
    def __init__(self, KO, arma, equipo_villanos):
        self.ko = KO
        self.arma = arma
        self.equipo_villanos = equipo_villanos
    
    def puntuacion(self):
            self.ko = False
            print("Uno de los villanos del Joker ha sido derrotado. Sigue matandolos!")
            self.equipo_villanos -= 1
            print("Quedan " ,self.equipo_villanos, " villanos por matar!\n\n" )

    def estado_armas_villanos(self):
        if(self.arma == True):
            print("El villano ha sido dañado por un PEM y esta en shok.")
        else:
            print("Tu PEM no ha alcanzado al villano ):")
```
